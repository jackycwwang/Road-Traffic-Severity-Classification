# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('RTA Dataset.csv')
```

```python
df.columns
```

```python
# show data types
df.dtypes
```

```python
# show number of null values in all columns
df.isnull().sum()
```

```python
# columns which have null values
[col for col in df.columns if df[col].isnull().sum() > 0]
```

```python
# visualize the null values
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df = df.drop(columns=['Service_year_of_vehicle', 'Defect_of_vehicle', 'Work_of_casuality', 'Fitness_of_casuality'])
```

```python
# no duplicated rows
df.duplicated().sum()
```

```python
df.Casualty_severity.value_counts()
```

```python
# It makes sense to replace 'na' with '0'
# Convert the data type to "int"
df.Casualty_severity = df.Casualty_severity.replace('na', '0').astype(int)
```

```python
df.Road_allignment.value_counts()
```

```python
df.Road_allignment.mode()
```

```python
# Replace columns with least null values with their respective mode value
df.Road_allignment = df.Road_allignment.fillna(df.Road_allignment.mode()[0])
df.Type_of_collision = df.Type_of_collision.fillna(df.Type_of_collision.mode()[0])
df.Road_surface_type = df.Road_surface_type.fillna(df.Road_surface_type.mode()[0])
df.Area_accident_occured = df.Area_accident_occured.fillna(df.Area_accident_occured.mode()[0])
df.Vehicle_movement = df.Vehicle_movement.fillna(df.Vehicle_movement.mode()[0])
```

```python
df.isnull().sum()
```

```python
df.describe()
```
